```python
font_size = 14
no_outline = {'ec': 'None'}

h = 3
w = 5

pgm = daft.PGM(grid_unit=3, node_unit=1.5, aspect=1, shape=[w,h])

pgm.add_node("pi", r"$\pi$", 2, 3, fontsize=font_size, plot_params=no_outline)
pgm.add_node("z", r"$z_i$", 2, 2, fontsize=font_size)
pgm.add_node("y", r"$y_i$", 2, 1, observed=True, fontsize=font_size)
pgm.add_node("mu", r"$\mu_k$", 3.5, 2, fontsize=font_size, plot_params=no_outline)
pgm.add_node("sigma", r"$\Sigma_k$", 3.5, 1, fontsize=font_size, plot_params=no_outline)

# pgm.add_node("x0", r"", 0, 0, fontsize=font_size)
# pgm.add_node("x1", r"", w, h, fontsize=font_size)

pgm.add_edge("mu", "y")
pgm.add_edge("sigma", "y")
pgm.add_edge("pi", "z")
pgm.add_edge("z", "y")

pgm.add_plate([1.5, 0.5, 1, 2], label=r"N", shift=-0.1, fontsize=font_size)
pgm.add_plate([3, 0.5, 1, 2], label=r"K", shift=-0.1, fontsize=font_size)

pgm.render()
# pgm.savefig('dirichlet-process-b.svg')
pgm.show()
```

```python
font_size = 14
no_outline = {'ec': 'None'}

h = 4
w = 5

pgm = daft.PGM(grid_unit=3, node_unit=1.5, aspect=1, shape=[w,h])

pgm.add_node("alpha", r"$\alpha$", 2, 3.9, fontsize=font_size, plot_params=no_outline)
pgm.add_node("pi", r"$\pi$", 2, 3, fontsize=font_size)
pgm.add_node("z", r"$z_i$", 2, 2, fontsize=font_size)
pgm.add_node("y", r"$y_i$", 2, 1, observed=True, fontsize=font_size)
pgm.add_node("mu", r"$\mu_k$", 3.5, 2, fontsize=font_size)
pgm.add_node("m", r"$m$", 4.5, 2.5, fontsize=font_size, plot_params=no_outline)
pgm.add_node("k", r"$\kappa$", 4.5, 2, fontsize=font_size, plot_params=no_outline)
pgm.add_node("sigma", r"$\Sigma_k$", 3.5, 1, fontsize=font_size)
pgm.add_node("nu", r"$\nu$", 4.5, 1, fontsize=font_size, plot_params=no_outline)
pgm.add_node("s", r"$S$", 4.5, 0.5, fontsize=font_size, plot_params=no_outline)

# pgm.add_node("x0", r"", 0, 0, fontsize=font_size)
# pgm.add_node("x1", r"", w, h, fontsize=font_size)

pgm.add_edge("mu", "y")
pgm.add_edge("m", "mu")
pgm.add_edge("k", "mu")
pgm.add_edge("sigma", "y")
pgm.add_edge("nu", "sigma")
pgm.add_edge("s", "sigma")
pgm.add_edge("alpha", "pi")
pgm.add_edge("pi", "z")
pgm.add_edge("z", "y")

pgm.add_plate([1.5, 0.5, 1, 2], label=r"N", shift=-0.1, fontsize=font_size)
pgm.add_plate([3, 0.5, 1, 2], label=r"K", shift=-0.1, fontsize=font_size)
pgm.add_plate([1.5, 2.7, 1, 0.7], label=r"K", shift=-0.1, fontsize=font_size)

pgm.render()
# pgm.savefig('dirichlet-process-b.svg')
pgm.show()
```

```python
font_size = 14

pgm = daft.PGM(grid_unit=3, node_unit=1.5, aspect=1, shape=[5,5])

# Hyperparameters
pgm.add_node("gamma", r"$\gamma$", 1, 3, fixed=True, fontsize=font_size)
pgm.add_node("H", r"$H(\lambda)$", 2, 4, fixed=True, fontsize=font_size)
pgm.add_node("G0", r"$G_0$", 2, 3, fontsize=font_size)
pgm.add_node("theta_prime", r"$\theta_i'$", 2, 2, fontsize=font_size)
pgm.add_node("y", r"$y_i$", 2, 1, observed=True, fontsize=font_size)


# Base distribution G0
pgm.add_edge("gamma", "G0")
pgm.add_edge("H", "G0")
pgm.add_edge("G0", "theta_prime")
pgm.add_edge("theta_prime", "y")

# And a plate.
pgm.add_plate([1.5, 0.5, 1, 2], label=r"$N$", shift=-0.1, fontsize=font_size)
# Observations
# Render and show the figure
pgm.render()
pgm.savefig('dirichlet-process-a.svg')
pgm.show()
```

```python
font_size = 14

pgm = daft.PGM(grid_unit=3, node_unit=1.5, aspect=1, shape=[5,5])

# Hyperparameters
pgm.add_node("gamma", r"$\gamma$", 1, 3, fixed=True, fontsize=font_size)
pgm.add_node("lambda", r"$\lambda$", 3.5, 3, fixed=True, fontsize=font_size)
pgm.add_node("theta", r"$\theta_k$", 3.5, 2, fontsize=font_size)
pgm.add_node("beta", r"$\beta$", 2, 3, fontsize=font_size)
pgm.add_node("z", r"$z_i$", 2, 2, fontsize=font_size)
pgm.add_node("y", r"$y_i$", 2, 1, observed=True, fontsize=font_size)


# Base distribution G0
pgm.add_edge("gamma", "beta")
pgm.add_edge("lambda", "theta")
pgm.add_edge("theta", "y")
pgm.add_edge("beta", "z")
pgm.add_edge("z", "y")

# And a plate.
pgm.add_plate([1.5, 0.5, 1, 2], label=r"$N$", shift=-0.1, fontsize=font_size)
pgm.add_plate([3, 1.5, 1, 1], label=r"$\infty$", shift=-0.1, fontsize=font_size+6)
# Observations
# Render and show the figure
pgm.render()
# pgm.savefig('dirichlet-process-b.svg')
pgm.show()
```
