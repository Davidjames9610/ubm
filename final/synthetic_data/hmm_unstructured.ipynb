```python
import numpy as np
import matplotlib.pyplot as plt
from hmmlearn import hmm

# Set random seed for reproducibility
np.random.seed(42)

# Number of states (clusters)
n_states = 3

# Create a GaussianHMM model
model = hmm.GaussianHMM(n_components=n_states, covariance_type="full")

# Set transition probabilities
model.transmat_ = np.array([[0.7, 0.2, 0.1],
                            [0.3, 0.5, 0.2],
                            [0.1, 0.3, 0.6]])

# Set initial state probabilities
model.startprob_ = np.array([0.6, 0.3, 0.1])

# Set means and covariances for each state
model.means_ = np.array([[0.0, 0.0], [5.0, 5.0], [10.0, 0.0]])
model.covars_ = np.tile(np.identity(2), (n_states, 1, 1))

# Generate sequences of observations
n_samples = 1000
X, Z = model.sample(n_samples)

# Plot the simulated data
plt.scatter(X[:, 0], X[:, 1], c=Z, cmap='viridis')
plt.title('Simulated Data from HMM with Overlapping Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
