```python
import librosa
import whale.setup.constants as const
import matplotlib.pyplot as plt
```

```python
whale_loc = const.ZOOM2_WAV
y, sr = librosa.load(whale_loc)
```

```python
t = 5 # seconds
s = int(t*sr)
audio = y[0:s]
plt.plot(audio)
```

```python
normalised = librosa.util.normalize(audio)
plt.plot(normalised)
```

```python
mfcc_feat = librosa.feature.mfcc(normalised, n_mfcc=13)
```

```python
import spafe.utils.vis as vis
```

```python
vis.show_features(mfcc_feat.T, 'mfcc', 'x', 'y')
```

```python
mfcc_feat_delta = librosa.feature.delta(mfcc_feat)
```

```python
vis.show_features(mfcc_feat_delta.T, 'mfcc_feat_delta', 'x', 'y')
```
